```python
def perceptron(x,y):
    coef=np.ones(x.shape[1])
    lr=0.01
    for i in range(1000):
        j=np.random.randint(0,x.shape[0])
        y_hat=(np.dot(x[j],coef))
        coef=coef -lr*(y[j]-y_hat)*x[j]
    return coef[0:],coef[1:]
```

```python
perceptron(x,y)
```

```python
coef_,intercept_=perceptron(x,y)
```
