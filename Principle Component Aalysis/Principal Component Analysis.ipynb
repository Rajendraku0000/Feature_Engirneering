```python
import numpy as np
```

```python
import pandas as pd
```

```python
np.random.seed(20)
val1=np.array([0,0,0])
val2=np.array([[1,0,0],[0,1,0],[0,0,1]])
new=np.random.multivariate_normal(val1,val2,20)
df=pd.DataFrame(new,columns=["column1","column2","column3"])
df["target"]=0
df
```

```python
val2=np.array([0,0,0])
val3=np.array([[1,0,0],[0,1,0],[0,0,1]])
new1=np.random.multivariate_normal(val2,val3,20)
df2=pd.DataFrame(new1,columns=["column1","column2","column3"])
df2["target"]=1
df2
```

```python
df=df.append(df2,ignore_index=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df.iloc[:,:3]=sc.fit_transform(df.iloc[:,:3])
```

```python
df.iloc[:,:3]
```

```python
np.round(df.describe())
```

```python
covariance_matrics=np.cov([df["column1"],df["column2"],df["column3"]])
```

```python
covariance_matrics
```

```python
eigen_values, eigen_vectors=np.linalg.eig(covariance_matrics)
```

```python
pc=eigen_vectors[:2]
pc
```

```python
df.iloc[:,:3]
```

```python
transformed=np.dot(df.iloc[:,:3],pc.T)
```

```python
new_data=pd.DataFrame(transformed,columns=["column1","columns2"])
```
